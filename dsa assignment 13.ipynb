Question 1

Given two linked list of the same size, the task is to create a new linked list using those linked lists. The condition is that the greater node among both linked list will be added to the new linked list.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
def merge_lists(list1, list2):
    ptr1 = list1
    ptr2 = list2
    newList = None
    newPtr = None

    while ptr1 and ptr2:
        if ptr1.data >= ptr2.data:
            if newList is None:
                newList = Node(ptr1.data)
                newPtr = newList
            else:
                newPtr.next = Node(ptr1.data)
                newPtr = newPtr.next
            ptr1 = ptr1.next
        else:
            if newList is None:
                newList = Node(ptr2.data)
                newPtr = newList
            else:
                newPtr.next = Node(ptr2.data)
                newPtr = newPtr.next
            ptr2 = ptr2.next

    # Append remaining nodes from list1, if any
    while ptr1:
        newPtr.next = Node(ptr1.data)
        newPtr = newPtr.next
        ptr1 = ptr1.next

    # Append remaining nodes from list2, if any
    while ptr2:
        newPtr.next = Node(ptr2.data)
        newPtr = newPtr.next
        ptr2 = ptr2.next

    return newList
     
list1 = Node(5)
list1.next = Node(2)
list1.next.next = Node(3)
list1.next.next.next = Node(8)

list2 = Node(1)
list2.next = Node(7)
list2.next.next = Node(4)
list2.next.next.next = Node(5)

result = merge_lists(list1, list2)

# Print the resulting linked list
while result:
    print(result.data, end=' ')
    result = result.next
```

Question 2


Write a function that takes a list sorted in non-decreasing order and deletes any duplicate nodes from the list. The list should only be traversed once.

For example if the linked list is 11->11->11->21->43->43->60 then removeDuplicates() should convert the list to 11->21->43->60.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def removeDuplicates(head):
    if head is None:
        return head

    current = head
    while current.next is not None:
        if current.data == current.next.data:
            current.next = current.next.next
        else:
            current = current.next

    return head

     

# Example usage:
# LinkedList: 11->11->11->21->43->43->60
head = Node(11)
head.next = Node(11)
head.next.next = Node(11)
head.next.next.next = Node(21)
head.next.next.next.next = Node(43)
head.next.next.next.next.next = Node(43)
head.next.next.next.next.next.next = Node(60)
result = removeDuplicates(head)

# Print the resulting linked list
while result is not None:
    print(result.data, end=' ')
    result = result.next
```

Given a linked list of size N. The task is to reverse every k nodes (where k is an input to the function) in the linked list. If the number of nodes is not a multiple of k then left-out nodes, in the end, should be considered as a group and must be reversed (See Example 2 for clarification).

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def reverseKNodes(head, k):
    if head is None:
        return None

    current = head
    next_node = None
    prev_node = None
    count = 0

    # Reverse the first k nodes of the linked list
    while current is not None and count < k:
        next_node = current.next
        current.next = prev_node
        prev_node = current
        current = next_node
        count += 1

    # Recursive call to reverse the next group of k nodes
    if next_node is not None:
        head.next = reverseKNodes(next_node, k)

    return prev_node
     
head = Node(1)
head.next = Node(2)
head.next.next = Node(2)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)

k = 4

result = reverseKNodes(head, k)

# Print the resulting linked list
while result is not None:
    print(result.data, end=' ')
    result = result.next
```

Question 5

Given a linked list and a key to be deleted. Delete last occurrence of key from linked. The list may have duplicates.


```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def deleteLastOccurrence(head, key):
    if head is None:
        return None

    # Keep track of the last occurrence and the previous node
    last_occurrence = None
    prev = None

    current = head
    prev_node = None

    # Traverse the linked list
    while current is not None:
        if current.data == key:
            last_occurrence = current
            prev = prev_node
        prev_node = current
        current = current.next

    # If last occurrence is found
    if last_occurrence is not None:
        # If it is the first node, update the head
        if prev is None:
            head = head.next
        else:
            prev.next = last_occurrence.next

    return head
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(5)
head.next.next.next.next = Node(2)
head.next.next.next.next.next = Node(10)

key = 2

result = deleteLastOccurrence(head, key)

# Print the resulting linked list
while result is not None:
    print(result.data, end=' ')
    result = result.next
```

Question 6


Given two sorted linked lists consisting of N and M nodes respectively. The task is to merge both of the lists (in place) and return the head of the merged list.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def mergeSortedLists(a, b):
    # If either list is empty, return the other list
    if a is None:
        return b
    if b is None:
        return a

    # Initialize pointers for the merged list and two input lists
    merged_head = None
    merged_tail = None
    ptr1 = a
    ptr2 = b

    # Set the head of the merged list to the smaller first node
    if ptr1.data <= ptr2.data:
        merged_head = ptr1
        ptr1 = ptr1.next
    else:
        merged_head = ptr2
        ptr2 = ptr2.next
    merged_tail = merged_head

    # Merge the lists by comparing and rearranging the pointers
    while ptr1 is not None and ptr2 is not None:
        if ptr1.data <= ptr2.data:
            merged_tail.next = ptr1
            ptr1 = ptr1.next
        else:
            merged_tail.next = ptr2
            ptr2 = ptr2.next
        merged_tail = merged_tail.next

    # Append any remaining nodes from the first list
    if ptr1 is not None:
        merged_tail.next = ptr1

    # Append any remaining nodes from the second list
    if ptr2 is not None:
        merged_tail.next = ptr2

    return merged_head

a = Node(5)
a.next = Node(10)
a.next.next = Node(15)

b = Node(2)
b.next = Node(3)
b.next.next = Node(20)

merged = mergeSortedLists(a, b)

# Print the merged list
while merged is not None:
    print(merged.data, end=' ')
    merged = merged.next
```

Question 7


Given a Doubly Linked List, the task is to reverse the given Doubly Linked List.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

def reverseDoublyLinkedList(head):
    if head is None:
        return None

    current = head

    # Swap prev and next pointers for each node
    while current is not None:
        temp = current.prev
        current.prev = current.next
        current.next = temp
        current = current.prev

    # Update the head
    head = temp.prev

    return head
head = Node(10)
node1 = Node(8)
node2 = Node(4)
node3 = Node(2)

head.next = node1
node1.prev = head
node1.next = node2
node2.prev = node1
node2.next = node3
node3.prev = node2

reversed_head = reverseDoublyLinkedList(head)

# Print the reversed linked list
current = reversed_head
while current is not None:
    print(current.data, end=' ')
    current = current.next
```

Question 8


Given a doubly linked list and a position. The task is to delete a node from given position in a doubly linked list.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

def deleteNode(head, position):
    if head is None:
        return None

    # If the position is 1, update the head
    if position == 1:
        head = head.next
        if head is not None:
            head.prev = None
        return head

    current = head
    count = 1

    # Traverse to the node at the given position
    while current is not None and count < position:
        current = current.next
        count += 1

    # If position is out of range, return the original list
    if current is None:
        return head

    # Update the previous and next pointers
    prev_node = current.prev
    next_node = current.next

    if prev_node is not None:
        prev_node.next = next_node
    if next_node is not None:
        next_node.prev = prev_node

    return head
     
head = Node(1)
node1 = Node(3)
node2 = Node(4)

head.next = node1
node1.prev = head
node1.next = node2
node2.prev = node1

position = 3

result = deleteNode(head, position)

# Print the resulting linked list
current = result
while current is not None:
    print(current.data, end=' ')
    current = current.next
     
```
